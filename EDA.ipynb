```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Churn_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.drop(columns=['customerID','SeniorCitizen','gender','Dependents','PaperlessBilling','PaymentMethod','MonthlyCharges','PhoneService','StreamingMovies'],axis=1)
```

```python
df.head()
```

```python
print("MultipleLines", df["MultipleLines"].unique())
print("Partner", df["Partner"].unique())
print("InternetService", df["InternetService"].unique())
print("OnlineSecurity", df["OnlineSecurity"].unique())
print("OnlineBackup", df["OnlineBackup"].unique())
print("DeviceProtection", df["DeviceProtection"].unique())
print("TechSupport", df["TechSupport"].unique())
print("StreamingTV", df["StreamingTV"].unique())
print("Contract", df["Contract"].unique())
print("Churn", df["Churn"].unique())
```

```python
df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
df['OnlineSecurity'] = df['OnlineSecurity'].replace('No internet service', 'No')
df['OnlineBackup'] = df['OnlineBackup'].replace('No internet service', 'No')
df['DeviceProtection'] = df['DeviceProtection'].replace('No internet service', 'No')
df['TechSupport'] = df['TechSupport'].replace('No internet service', 'No')
df['StreamingTV'] = df['StreamingTV'].replace('No internet service', 'No')
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
label=LabelEncoder()
encoder=OneHotEncoder(sparse_output=False)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
x_train['Partner']=label.fit_transform(x_train['Partner'])
x_train['StreamingTV']=label.fit_transform(x_train['StreamingTV'])
x_train['TechSupport']=label.fit_transform(x_train['TechSupport'])
x_train['OnlineBackup']=label.fit_transform(x_train['OnlineBackup'])
x_train['OnlineSecurity']=label.fit_transform(x_train['OnlineSecurity'])
```

```python
x_train['DeviceProtection']=label.fit_transform(x_train['DeviceProtection'])
x_train['MultipleLines']=label.fit_transform(x_train['MultipleLines'])
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# One encoder for InternetService
encoder_internet = OneHotEncoder(sparse_output=False)
Internet_train_encoded = encoder_internet.fit_transform(x_train[['InternetService']])
Internet_train_encoded_df = pd.DataFrame(
    Internet_train_encoded,
    columns=encoder_internet.get_feature_names_out(['InternetService']),
    index=x_train.index
)
x_train = pd.concat([x_train, Internet_train_encoded_df], axis=1).drop(columns=['InternetService'])

# One encoder for Contract
encoder_contract = OneHotEncoder(sparse_output=False)
Contract_train_encoded = encoder_contract.fit_transform(x_train[['Contract']])
Contract_train_encoded_df = pd.DataFrame(
    Contract_train_encoded,
    columns=encoder_contract.get_feature_names_out(['Contract']),
    index=x_train.index
)
x_train = pd.concat([x_train, Contract_train_encoded_df], axis=1).drop(columns=['Contract'])
```

```python
x_train['TotalCharges'] = pd.to_numeric(x_train['TotalCharges'], errors='coerce')
x_test['TotalCharges']  = pd.to_numeric(x_test['TotalCharges'], errors='coerce')
mean_total_charges = x_train['TotalCharges'].mean()
x_train['TotalCharges'].fillna(mean_total_charges, inplace=True)
x_test['TotalCharges'].fillna(mean_total_charges, inplace=True)
```

```python
x_train.head()
```

```python
# Apply label encoding with the SAME label encoder
x_test['Partner'] = label.transform(x_test['Partner'])
x_test['StreamingTV'] = label.transform(x_test['StreamingTV'])
x_test['TechSupport'] = label.transform(x_test['TechSupport'])
x_test['OnlineBackup'] = label.transform(x_test['OnlineBackup'])
x_test['OnlineSecurity'] = label.transform(x_test['OnlineSecurity'])
x_test['DeviceProtection'] = label.transform(x_test['DeviceProtection'])
x_test['MultipleLines'] = label.transform(x_test['MultipleLines'])
```

```python
# InternetService
Internet_test_encoded = encoder_internet.transform(x_test[['InternetService']])
Internet_test_encoded_df = pd.DataFrame(
    Internet_test_encoded,
    columns=encoder_internet.get_feature_names_out(['InternetService']),
    index=x_test.index
)
x_test = pd.concat([x_test, Internet_test_encoded_df], axis=1).drop(columns=['InternetService'])

# Contract
Contract_test_encoded = encoder_contract.transform(x_test[['Contract']])
Contract_test_encoded_df = pd.DataFrame(
    Contract_test_encoded,
    columns=encoder_contract.get_feature_names_out(['Contract']),
    index=x_test.index
)
x_test = pd.concat([x_test, Contract_test_encoded_df], axis=1).drop(columns=['Contract'])
```

```python
y_train=label.fit_transform(y_train)
```

```python
y_test=label.transform(y_test)
```

```python
x_train['TotalCharges'] = pd.to_numeric(x_train['TotalCharges'], errors='coerce')
x_test['TotalCharges']  = pd.to_numeric(x_test['TotalCharges'], errors='coerce')
```

```python
from sklearn.preprocessing import StandardScaler
Sclar=StandardScaler()
```

```python
x_train=Sclar.fit_transform(x_train)
x_test=Sclar.transform(x_test)
```
